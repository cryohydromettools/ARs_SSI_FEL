```python
import pandas as pd
import xarray as xr
```

```python
filename = 'data/EE_SSI_Florencia.txt'
EE_SSI = pd.read_csv(filename, sep='\t')

EE_SSI['date'] = pd.to_datetime(EE_SSI[['year', 'month', 'day']])

EE_SSI
```

```python
filename = 'data/Z500_index_daily.csv'
Z500_index = pd.read_csv(filename, sep='\t')
Z500_index['date'] = pd.to_datetime(Z500_index['time'])
Z500_index
```

```python
filename = 'data/norm.daily.aao.cdas.z700.19790101_current.csv'
SAM_index = pd.read_csv(filename, sep=',')
SAM_index['date'] = pd.to_datetime(SAM_index[['year', 'month', 'day']])
SAM_index
```

```python
filename = 'data/SSI_ERA5_optm_19690101-20200331.nc'
ds = xr.open_dataset(filename).isel(lat=2, lon=2)[['RAIN', 'SNOWFALL', 'surfM', 'Q']].resample(time='1d').sum()*1000
#ds.sel(lon=-58.886878, lat=-62.154407, method='nearest')#[['RAIN', 'SNOWFALL', 'surfM', 'Q']]
ds
```

```python
df_bell = ds.to_dataframe().reset_index(drop=False).rename(columns={'time':'date'})[['date', 'surfM', 'Q']]
```

```python
EE_SSI = pd.merge(EE_SSI, df_bell, on='date', how='left')
```

```python
EE_Z500 = pd.merge(EE_SSI, Z500_index, on='date', how='left')
EE_Z500
```

```python
filename = '/home/cr2/cmtorres/repos/ARs_SSI_FEL/data/ARs_Guan_AP_day.csv'
ARs_fall = pd.read_csv(filename, sep='\t', parse_dates=['date'])
ARs_fall
```

```python
EE_Z500_ARs = pd.merge(EE_SSI, ARs_fall, on='date', how='left')
EE_Z500_ARs
```

```python
EE_Z500 = pd.merge(EE_Z500_ARs, Z500_index[['date', 'z_anomaly']], on='date', how='left')
EE_Z500_SAM = pd.merge(EE_Z500, SAM_index[['date', 'aao_index_cdas']], on='date', how='left')
vars = ['date', 'aao_index_cdas', 'z_anomaly',  'ff (> 2sd)', 'pp (> 3sd) ', 'T_BCAA', 'T_FREI', 'T_KOREA', 'surfM', 'Q', 'ARs', 'date.1']
EE_Z500_SAM = EE_Z500_SAM[vars]
EE_Z500_SAM
```

```python
EE_Z500_SAM.dropna().to_csv('data/ARs_Guan_EMEs_SSI.csv', sep='\t', index=False)
EE_Z500_SAM.dropna()
```

```python
filename = '~/ARs_DATA/Global_AR_dataset_Guan.nc'
ds = xr.open_dataset(filename)
de_sel = ds[['shapemap']].sel(time = EE_Z500_SAM.dropna()['date.1'].values, ens = 1, lev = 0)
de_sel.coords['lon'] = (de_sel.coords['lon'] + 180) % 360 - 180
de_sel = de_sel.sortby(de_sel.lon)
de_sel = de_sel.sel(lon=slice(-150, -30), lat=slice(-30, -75))
de_sel = de_sel.sel(time=de_sel['time'].dt.month.isin([11, 12, 1, 2, 3]))
```

```python
de_sel.to_netcdf('data/ARs_Guan_HW_SSI.nc')
```

```python
E_Z500_SAM_1999 = EE_Z500_SAM[(EE_Z500_SAM.date >= '1999-11-12') & (EE_Z500_SAM.date <= '1999-11-20')]
E_Z500_SAM_1999

# Supondo que seu DataFrame já está carregado como E_Z500_SAM_2005
df = E_Z500_SAM_1999.copy()
E_Z500_SAM_1999['date'] = E_Z500_SAM_1999['date'].dt.date
# Colunas que serão **somadas**
cols_sum = ['pp (> 3sd) ', 'surfM', 'Q']

# Colunas que terão **média**
cols_mean = ['aao_index_cdas', 'z_anomaly', 'ff (> 2sd)', 'T_BCAA', 'T_FREI', 'T_KOREA']

# Criar série com os valores médios e somados
mean_values = df[cols_mean].mean()
sum_values = df[cols_sum].sum()

# Combinar os resultados
summary = pd.concat([mean_values, sum_values])

# Adicionar uma data representativa
summary['date'] = 'Mean/Sum'
summary['ARs'] = '-'
summary['date.1'] = '-'

# Reordenar as colunas para combinar com o DataFrame original
summary = summary[df.columns]  # ordenar como no original
df_summary = pd.DataFrame([summary])
df_final = pd.concat([E_Z500_SAM_1999, df_summary], axis=0)

# Se quiser ver o resultado
(df_final)
```
