```python
import pandas as pd
import xarray as xr
```

```python
filename = 'data/EE_SSI_Florencia.txt'
EE_SSI = pd.read_csv(filename, sep='\t')

EE_SSI['date'] = pd.to_datetime(EE_SSI[['year', 'month', 'day']])

EE_SSI
```

```python
filename = 'data/Z500_index_daily.csv'
Z500_index = pd.read_csv(filename, sep='\t')
Z500_index['date'] = pd.to_datetime(Z500_index['time'])
Z500_index
```

```python
filename = 'data/norm.daily.aao.cdas.z700.19790101_current.csv'
SAM_index = pd.read_csv(filename, sep=',')
SAM_index['date'] = pd.to_datetime(SAM_index[['year', 'month', 'day']])
SAM_index
```

```python
filename = 'data/SSI_ERA5_optm_19690101-20200331.nc'
ds = xr.open_dataset(filename).isel(lat=2, lon=2)[['RAIN', 'SNOWFALL', 'surfM', 'Q']].resample(time='1d').sum()*1000
#ds.sel(lon=-58.886878, lat=-62.154407, method='nearest')#[['RAIN', 'SNOWFALL', 'surfM', 'Q']]
ds
```

```python
df_bell = ds.to_dataframe().reset_index(drop=False).rename(columns={'time':'date'})[['date', 'surfM', 'Q']]
```

```python
EE_SSI = pd.merge(EE_SSI, df_bell, on='date', how='left')
```

```python
EE_Z500 = pd.merge(EE_SSI, Z500_index, on='date', how='left')
EE_Z500
```

```python
EE_Z500 = pd.merge(EE_SSI, Z500_index[['date', 'z_anomaly']], on='date', how='left')
EE_Z500_SAM = pd.merge(EE_Z500, SAM_index[['date', 'aao_index_cdas']], on='date', how='left')
vars = ['date', 'aao_index_cdas', 'z_anomaly',  'ff (> 2sd)', 'pp (> 3sd) ', 'T_BCAA', 'T_FREI', 'T_KOREA', 'surfM', 'Q']
EE_Z500_SAM = EE_Z500_SAM[vars]
EE_Z500_SAM
```

```python
E_Z500_SAM_1999 = EE_Z500_SAM[(EE_Z500_SAM.date >= '1999-11-12') & (EE_Z500_SAM.date <= '1999-11-20')]
E_Z500_SAM_1999
```

```python
E_Z500_SAM_1999.agg({'aao_index_cdas': 'mean', 'z_anomaly': 'mean', 'ff (> 2sd)': 'mean', 'pp (> 3sd) ': 'sum', 
                     'T_BCAA': 'mean', 'T_FREI': 'mean', 'T_KOREA': 'mean', 'surfM': 'sum', 'Q': 'sum',})
```

```python
E_Z500_SAM_2005 = EE_Z500_SAM[(EE_Z500_SAM.date >= '2005-02-09') & (EE_Z500_SAM.date <= '2005-02-20')]
E_Z500_SAM_2005
```
